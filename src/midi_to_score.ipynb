```python
from music21 import converter, stream, environment
import pathlib

environment.set('musescoreDirectPNGPath', '/opt/homebrew/bin/mscore')

def convert_midi_to_sheet(midi_file_path, output_file_path):
    # Load MIDI file
    midi_stream = converter.parse(midi_file_path)

    # Create a new stream for notation
    score = stream.Score()

    # Iterate through each instrument part in the MIDI file
    for part in midi_stream.parts:
        # Create a new part in the score
        score_part = stream.Part()

        # Iterate through each note in the part
        for element in part.notes:
            # Add the note to the part in the score
            score_part.append(element)

        # Add the part to the score
        score.append(score_part)

    # Write the score to a new file
    score.write('xml', fp=output_file_path)

# Example usage
midi_file_path = 'data/midis/Aaron, Michael, Piano Course, V8WvKK-1b2c.mid'
output_file_path = 'data/output/Aaron, Michael, Piano Course, V8WvKK-1b2c.xml'
convert_midi_to_sheet(midi_file_path, output_file_path)
```
