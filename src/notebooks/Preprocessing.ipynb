```python
path = "/Users/yleniafedi/code/piano_transcription/piano_transcription/midis"
files = list(os.scandir(path))

#using a fourier transform from librosa - stft

result_stft = {}
for file in files[0:4]:

    soundfont = '/Users/yleniafedi/code/piano_transcription/piano_transcription/FluidR3_GM/FluidR3_GM.sf2'
    fs = FluidSynth(soundfont)

    fs.midi_to_audio(file, f'{file.name}.wav')
    output = f'{file.name}.wav'

    y, sr = librosa.load(output, duration=15, sr=None)
    S = np.abs(librosa.stft(y))

    #using a fourier transform from librosa - stft
    fig, ax = plt.subplots()
    img = librosa.display.specshow(librosa.amplitude_to_db(S,
                                                       ref=np.max),
                               y_axis='log', x_axis='time', ax=ax)
    ax.set_title('Power spectrogram')
    fig.colorbar(img, ax=ax, format="%+2.0f dB")

    result_stft[file.name] = fig.colorbar(img, ax=ax, format="%+2.0f dB")
```
